Python Match = Multiple conditions handle krne ka clean trikaa

```python
days = 1

match days:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3 :
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case 6:
        print("Saturday")
    case 7:
        print("Sunday") 
```

Default value = 
Default Value (case _:) in Match-Case

case _: default case hota hai.

Ye tab chalta hai jab upar ke kisi bhi case se value match nahi hoti.

Isko else ki tarah samajh sakte ho.

```python
Month = 49

match Month:
    case 1:
        print("January")
    case 2:
        print("February")
    case 3:
        print("March")
    case 4:
        print("April")
    case 5:
        print("May")
    case 6:
        print("June")
    case 7:
        print("July")
    case 8:
        print("August")
    case 9:
        print("September")
    case 10:
        print("October")
    case 11:
        print("November")
    case _:
        print("Tere baap ne 12 se jyada mahine banaye kya?")
  
```

Combine values = Agar ek hi code multiple values ke liye chalana ho, to values ko | (pipe operator) se combine kar sakte ho.

```python
Day = 6

match Day:
    case 1 | 2 | 3 | 4 | 5:
        print("Weekday")

    case 6 | 7:
        print("Weekend")
```

```python
month = 2

match month:
    case 12 | 1 | 2:
        print("Winter")
    case 3 | 4 | 5:
        print("Summer")
    case _:
        print("Other Season")
```

Recursion = Recursion in Python is a technique where a function calls itself to solve a problem by breaking it into smaller subproblems.

Recursion me 2 important cases hote hain:

1. Base Case = Ye recursion ko rokta hai.

2. Recursive Case = Ye function ko dobara call karta hai.

```python
def count(n):
    if n == 0:                  # Base Case
        return
    
    print(n)
    count(n - 1)                # Recursice case

count(5)
```

```python
def factorial(n):
    if n == 0 or  n == 1:
        return 1
    
    else:
        return n * factorial (n - 1)
    
print(factorial(5))
```

send() Method = The send() method allows you to send a value to the generator:

```python
def echo_generator():
  while True:
    received = yield
    print("Received:", received)

gen = echo_generator()
next(gen) # Prime the generator
gen.send("Hello")
gen.send("World")
```

close() Method = The close() method stops the generator:

```python
def my_gen():
  try:
    yield 1
    yield 2
    yield 3
  finally:
    print("Generator closed")

gen = my_gen()
print(next(gen))
gen.close()
```

Create a Child class = To create a class that inherits the functionality from another class, send the parent class as a parameter when creating a child class.

In simple language create a child class ka matlab hai ek nayi class banana jo kisi dusri class (Parent Class) ki properties aur methods ko inherit kare.

```python
class Person:
    def __init__(self, fname, lname):
        self.firstname = fname
        self.lastname = lname

    def printname(self):
        print(self.firstname, self.lastname)

class Student(Person):
    pass

x = Student("Mike", "Olsen")
x.printname()
```

So far we have created a child class that inherits the properties and methods from its Parent.

Add the __init__() function to child class = We want to add the __init__() function to the child class.

1. When you add the __init__() function the child class will no longer inherits the parents __init__() function.

2. NOTE = The childs __init__() function overrides the inheritance of the parents __init__() function.

```python
class Person:
    def __init__(self, fname, lname):
        self.firstname = fname
        self.lastname = lname
    def printname(self):
        print(self.firstname, self.lastname)

class Student(Person):
    def __init__(self, fname, lname):           # To keep the __init__ function of parent class, add a call to the parents __init__ function
        Person.__init__(self, fname, lname)

x = Student("Mike", "Olsen")
x.printname()
```

Super function = super() ka use parent class ke methods aur constructor ko call karne ke liye hota hai.

```python
class Person:
    def __init__(self, fname, lname):
        self.firstname = fname
        self.lastname = lname

    def printname(self):
        print(self.firstname, self.lastname)

class Student(Person):
    def __init__(self, fname, lname):
        super().__init__(fname, lname)     # Here we can use Person.__init__ too

x = Student("Bhagvat", "Narendra")
x.printname()
```

Python Polymorphism = The word Polymorphism means many forms and in programming it refers to Methods/Functions/Operators with the same name that can be executed on many objects or classes.

Polymorphism allows the same method, function, or operator name to be used for different types of objects, with each object responding in its own way.

Programming mein iska matlab hai ki same method/function/operator ka naam ek hi ho sakta hai, lekin alag-alag objects ya classes par use karne par alag kaam karta hai.

Poly = Many

Morphism = Forms

```python
class Dog:
    def sound(self):
        print("Bark")

class Cat:
    def sound(self):
        print("Meow")

class Boat:
    def sound(self):
        print("Sail")

d = Dog()
c = Cat()
b = Boat()

d.sound()
c.sound()
b.sound()
```
